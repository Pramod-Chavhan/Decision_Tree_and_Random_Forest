Decision Tree is a Supervised learning technique that can be used for both classification and Regression problems.
Decision Tree uses a flowchart like a tree structure to show the predictions that result from a series of feature-based splits. It starts with a root node and ends with a decision made by leaves.

Random Forest
A random forest is a machine learning technique that’s used to solve regression and classification problems. It utilizes ensemble learning, which is a technique that combines many classifiers to provide solutions to complex problems.

A random forest algorithm consists of many decision trees. The ‘forest’ generated by the random forest algorithm is trained through bagging or bootstrap aggregating.

Desicion Tree 4. In the recruitment domain, HR faces the challenge of predicting if the candidate is faking their salary or not. For example, a candidate claims to have 5 years of experience and earns 70,000 per month working as a regional manager. The candidate expects more money than his previous CTC. We need a way to verify their claims (is 70,000 a month working as a regional manager with an experience of 5 years a genuine claim or does he/she make less than that?) Build a Decision Tree and Random Forest model with monthly income as the target variable.

Business objective: predict whether the candidate is faking his salary or not based on experiance.

```python
# Importing Fraud_check dataset using pandas
import pandas as pd
hr = pd.read_csv(r"C:\Users\D\Desktop\New Assignments  Keys\Datasets\HR_DT.csv")
```

```python
hr.columns
```

```python
### EDA, Pre-processing
hr.info()
hr.head()
```

monthly income is the target variable which we have to predict

other 2 variables are predictore - no. of years is numerical

position is categorical which has to be converted to numeric

```python
# Missing values
hr.isnull().sum()
# there are no missing values in our dataset
```

```python
hr['Position of the employee'].value_counts(normalize = True)
# we see this is ordinal data so we can convert it to numerical position wise, higher position - higher number, lower position - lower number
```

```python
position_label_mapping = {'Business Analyst': 1, 'Junior Consultant': 2, 'Senior Consultant': 3, 'Manager': 4, 'Country Manager': 5,'Region Manager': 6,
                          'Partner': 7, 'Senior Partner': 8, 'C-level': 9, 'CEO': 10}

hr['Position of the employee'] = hr['Position of the employee'].map(position_label_mapping)
```

```python
# Input and Output Split
predictors = hr.iloc[:,:-1]
target = hr[" monthly income of employee"]
```

```python
# Splitting data into training and testing data set
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(predictors, target, test_size = 0.25, random_state=0)
```

```python
from sklearn import tree
regtree = tree.DecisionTreeRegressor()
regtree.fit(x_train, y_train)
```

```python
# Prediction
test_pred = regtree.predict(x_test)
train_pred = regtree.predict(x_train)
```

```python
# Measuring accuracy
from sklearn.metrics import mean_squared_error, r2_score
# Error on test dataset
mean_squared_error(y_test, test_pred)
```

```python
r2_score(y_test, test_pred)
# we get Rsquare of 0.98 on test dataset which is very good
```

```python
# Error on train dataset
mean_squared_error(y_train, train_pred)
```

```python
r2_score(y_train, train_pred)
# we get Rsquare of 0.999 which is almost 1 on train dataset
```

```python
# let us try Random forest
from sklearn.ensemble import RandomForestRegressor
rf_clf = RandomForestRegressor(n_estimators=500, n_jobs=1, random_state=42)
rf_clf.fit(x_train, y_train)
```

```python
# Error on test dataset
mean_squared_error(y_test, rf_clf.predict(x_test))
```

```python
r2_score(y_test, rf_clf.predict(x_test))
# we see that our mean squared error goes down
# we also see that our R squared is now 99.15% which has increased
# we will therefore use this model generated by using RandomForrest
```

```python
# let us now check if candidate is honest or fraud
# first we will have to add his entry to our x_test data
cand = {'Position of the employee': 6, 'no of Years of Experience of employee': 5.0}
x_test_cand = x_test.append(cand, ignore_index = True)
x_test_preds = rf_clf.predict(x_test_cand)
# since it was the last row that we added, we are only interested in predicted value of the last row
```
